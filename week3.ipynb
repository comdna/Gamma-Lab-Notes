在这项工作中，您需要使用基于numpy的机器学习方法实现简单的线性回归。最好不要使用PyTorch或Tensorflow等深度学习库。

The first is the data generation. We can generate data that follows a linear distribution with the following functions.
首先是数据生成。我们可以用下列函数生成遵循线性分布的数据。

```python
import numpy as np
import matplotlib.pyplot as plt

def get_data(num):
  for _ in range(num):
    x = np.random.uniform(-10.0, 10.0)
    noise = np.random.normal(0, 1)
    y = x * 2 + 1 + noise
    yield np.array([x]).astype(np.float32), np.array([y]).astype(np.float32)

x = np.array([])
y = np.array([])
num=1000
for x0, y0 in get_data(num):
  x = np.append(x, x0)
  y = np.append(y, y0)

plt.scatter(x, y, color='blue', s=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Data Points')
plt.grid(True)
plt.show()
```

The following is setting the hyperparameters, and the initialization of the learnable parameters.  
下面是设置超参数和初始化可学习参数。

```python
learn_rate = 0.0001
epochs = 200
display_step = 50
np.random.seed(0)
w = np.random.rand()
b = np.random.rand()
```

So, you only need to write the code to train the learnable parameters w and b. We recommend you use the mean square error function.  
因此，您只需要编写代码来训练可学习的参数w和b。我们建议您使用均方误差函数。

```python
# 训练模型
for epoch in range(epochs):
    total_loss = 0.0
    for x0, y0 in get_data(num):
        # 前向传播
        y_pred = w * x0 + b
        
        # 计算损失函数
        loss = np.square(y_pred - y0).mean()
        total_loss += loss
        
        # 反向传播，更新参数
        w_gradient = 2 * np.mean((y_pred - y0) * x0)
        b_gradient = 2 * np.mean(y_pred - y0)
        w -= learn_rate * w_gradient
        b -= learn_rate * b_gradient
    
    total_loss /= num
    
    # 打印训练信息
    if (epoch + 1) % display_step == 0:
        print("Epoch:", '%d' % (epoch + 1), "loss=", "{:.9f}".format(total_loss))

print("w =", w)
print("b =", b)
```

```python
# 生成拟合线上的点
x_fit = x
m=x.shape[0]
y_fit = w * x_fit + b

# 绘制原始数据点和拟合线

plt.scatter(x, y, c='b',label='Original Data', s=5)
plt.plot(x_fit, y_fit, c='r', label='Fitted Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
