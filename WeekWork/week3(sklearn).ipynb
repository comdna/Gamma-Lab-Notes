The first is the data generation. We can generate data that follows a linear distribution with the following functions.
首先是数据生成。我们可以用下列函数生成遵循线性分布的数据。

```python
import numpy as np
import matplotlib.pyplot as plt

def get_data(num):
    data = []
    for _ in range(num):
        x = np.random.uniform(-10.0, 10.0)
        noise = np.random.normal(0, 1)
        y = x * 2 + 1 + noise
        data.append((np.array([[x]], dtype=np.float32), np.array([y], dtype=np.float32)))
    return data

num = 100
data = get_data(num)

X = np.concatenate([x for x, _ in data])  # 将列表中的数组连接成二维数组
y = np.concatenate([y for _, y in data])  # 将列表中的数组连接成一维数组

print("X shape:", X.shape)
print("y shape:", y.shape)


plt.scatter(X, y, color='blue', s=5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of Data Points')
plt.grid(True)
plt.show()

print(X)
print(y)
```
